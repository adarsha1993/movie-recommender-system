```python
import pandas as pd

# Load all datasets
credits       = pd.read_csv('../data/raw/credits.csv')
keywords      = pd.read_csv('../data/raw/keywords.csv')
links         = pd.read_csv('../data/raw/links.csv')
links_small   = pd.read_csv('../data/raw/links_small.csv')
metadata      = pd.read_csv('../data/raw/movies_metadata.csv', low_memory=False)
ratings_small = pd.read_csv('../data/raw/ratings_small.csv')

# Verify shapes
datasets = {
    'credits': credits, 'keywords': keywords,
    'links': links, 'links_small': links_small,
    'metadata': metadata, 'ratings_small': ratings_small
}

for name, df in datasets.items():
    print(f"{name:20s} → {df.shape[0]:>7,} rows | {df.shape[1]} cols | columns: {list(df.columns)}")
```
